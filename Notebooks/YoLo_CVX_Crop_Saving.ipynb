```python
model = torch.hub.load(current_path , 'custom', path = 'D:/CVX/Cervix_Cancer_Classification/exp2/weights/best.pt',source='local' )
```

```python
type1_original_filenames = os.listdir("D:\\CVX\\Cervix_Cancer_Classification\\train\\Type_1")
type2_original_filenames = os.listdir("D:\\CVX\\Cervix_Cancer_Classification\\train\\Type_2")
type3_original_filenames = os.listdir("D:\\CVX\\Cervix_Cancer_Classification\\train\\Type_3")
```

```python
for i in tqdm(range(len(type1_original_filenames))):
  path = 'D:/CVX/Cervix_Cancer_Classification/train/Type_1' + '/' + type1_original_filenames[i]
  img = path
  results = model(img)
  crop = results.crop(save=True, save_dir='D:\\CVX\\Cervix_Cancer_Classification\\train_processed\\Type_1', exist_ok=True)
```
